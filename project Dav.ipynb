Filling Null Values

```python
Data1["Genre"].fillna(Data1["Genre"].mode()[0],inplace=True)
Data1["Genre"].isnull().sum()
```

```python
NaS_mode = Data1["NA_Sales"].mean()
Data1["NA_Sales"].fillna(NaS_mode,inplace=True)
Data1["NA_Sales"].isnull().sum()
```

```python
Data1["Publisher"].mode()[0]
Data1["Publisher"].fillna("Electronic Arts",inplace=True)
```

```python
Data1["Year"].fillna(Data1["Year"].mode()[0],inplace=True)
Data1["Year"].isnull().sum()
```

```python
Data1["Year"]=pd.to_numeric(Data1["Year"])
```

```python
Data1.isnull().sum()
```

```python
Data1.duplicated().sum()
```

```python
Data1["Index"] = np.arange(len(Data1))
```

```python
Data1.set_index("Index",inplace=True)
```

```python
Data1["Genre"].value_counts()
```

```python
Data1["Rated"]=Data1["Genre"]
```

```python
Data1["Rated"]=Data1["Rated"].replace({"Action":"12+","Shooter":"12+","Fighting":"12+"})
Data1["Rated"]=Data1["Rated"].replace({"Sports":"Any Age Group","Platform":"Any Age Group","Racing":"Any Age Group","Role-Playing":"Any Age Group","Simulation":"Any Age Group","Adventure":"Any Age Group","Strategy":"Any Age Group","Puzzle":"Any Age Group","Misc":"Any Age Group"})
```

```python
v1 = Data1["Genre"].value_counts() 
v2 = Data1["Genre"].unique()
plt.figure(figsize=(12,6))
plt.pie(v1,labels=v2,autopct="%1.1f%%",shadow=True)
plt.show()
```

```python
b = Data1.groupby(["Genre"])["Global_Sales"].max()
plt.figure(figsize=(12,10))
sns.jointplot(b,kind="scatter",color="red")
```

```python
c = Data1.groupby(["Genre"])["Publisher"].count()
plt.figure(figsize=(12,5))
sns.barplot(c,color="blue")
plt.grid(True)
```

```python
g = Data1["Publisher"].value_counts()
g = g[:10]
g.plot(kind="bar",color="purple")
plt.title("TOP 10 PUBLISHERS")
```

```python
e = Data1.sort_values(["NA_Sales"],ascending=False)[["Name","NA_Sales"]]
e = e[:10].set_index("Name")
e.plot(kind="bar",color="blue")
plt.title("Top 10 North America VG Sales")
```

```python
f = Data1.sort_values(["JP_Sales"],ascending=False)[["Name","JP_Sales"]]
f = f[:10].set_index("Name")
f.plot(kind="bar",color='red')
plt.title("Top 10 Japan VG Sales")
```
